The center of the rangoli has the first alphabet letter a, and the boundary has the $N^{th}$ alphabet letter (in alphabetical order).

## Function Description
Complete the rangoli function in the editor below.<br>
rangoli has the following parameters:<br>
* int size: the size of the rangoli

### Returns
* string: a single string made up of each of the lines of the rangoli separated by a newline character (\n)

### Input Format
Only one line of input containing **size**, the size of the rangoli.

### Constraints
$ 0 < size < 27 $



```python
def print_rangoli(size):
    w = size*4-3
    line = ""
    # upper part
    for i in range(1,size+1):
        for j in range(0,i):
            line += chr(96+size-j)
            if len(line) < w :
                line += '-'
        for k in range(i-1,0,-1):    
            line += chr(97+size-k)
            if len(line) < w :
                line += '-'
        print(line.center(w,'-'))
        line = ''
        
    #lower part
    for i in range(size-1,0,-1):
        line = ''
        for j in range(0,i):
            line += chr(96+size-j)
            if len(line) < w :
                line += '-'
        for k in range(i-1,0,-1):
            line += chr(97+size-k)
            if len(line) < w :
                line += '-'
        print(line.center(w,'-'))
    
if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
```
