# Text Alignment
In Python, a string of text can be aligned left, right and center.<br>
**.ljust(width)**<br>
This method returns a left aligned string of length width.<br>
> `>>> width = 20`<br>
`>>> print 'HackerRank'.ljust(width,'-')`<br>
`HackerRank----------`<br>

**.center(width)**<br>
This method returns a centered string of length width.<br>
> `>>> width = 20`<br>
`>>> print 'HackerRank'.center(width,'-')`<br>
`-----HackerRank-----`<br>

**.rjust(width)**<br>
This method returns a right aligned string of length width.<br>
> `>>> width = 20`<br>
`>>> print 'HackerRank'.rjust(width,'-')`<br>
`----------HackerRank`<br>
### Task
You are given a partial code that is used for generating the HackerRank Logo of variable thickness.<br>
Your task is to replace the blank (______) with rjust, ljust or center.<br>
### Input Format
A single line containing the thickness value for the logo.
### Constraints
The thickness must be an odd number.<br>
$ 0 < thickness < 50 $
### Output Format
Output the desired logo.

```python
#Replace all ______ with rjust, ljust or center. 

thickness = int(input()) #This must be an odd number
c = 'H'

#Top Cone
for i in range(thickness):
    print((c*i).rjust(thickness-1)+c+(c*i).ljust(thickness-1))

#Top Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))

#Middle Belt
for i in range((thickness+1)//2):
    print((c*thickness*5).center(thickness*6))    

#Bottom Pillars
for i in range(thickness+1):
    print((c*thickness).center(thickness*2)+(c*thickness).center(thickness*6))    

#Bottom Cone
for i in range(thickness):
    print(((c*(thickness-i-1)).rjust(thickness)+c+(c*(thickness-i-1)).ljust(thickness)).rjust(thickness*6))
```
