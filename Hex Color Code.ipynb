# Hex Color Code
CSS colors are defined using a hexadecimal (HEX) notation for the combination of Red, Green, and Blue color values (RGB).<br>

Specifications of HEX Color Code<br>

■ It must start with a '#' symbol.<br>
■ It can have $3$ or $6$ digits.<br>
■ Each digit is in the range of $0$ to $F$. ($1,2,3,4,5,6,7,8,9,A,B,C,D,E$ and $F$).<br>
■ $A-F$ letters can be lower case. ($a,b,c,d,e$ and $f$ are also valid digits).<br>

### Examples

> `Valid Hex C<br>olor Codes`<br>
`#FFF `<br>
`#025 `<br>
`#F0A1FB `<br>

`Invalid Hex Color Codes`<br>
`#fffabg`<br>
`#abcf`<br>
`#12365erff`<br>

You are given $N$ lines of CSS code. Your task is to print all valid Hex Color Codes, in order of their occurrence from top to bottom.<br>

CSS Code Pattern<br>

> `Selector`<br>
`{`<br>
`	Property: Value;`<br>
`}`<br>

### Input Format

The first line contains $N$, the number of code lines.
The next $N$ lines contains CSS Codes.

### Constraints

$ 0 < N < 50 $

### Output Format

Output the color codes with '#' symbols on separate lines.

### Sample Input

> `11`<br>
`#BED`<br>
`{`<br>
`    color: #FfFdF8; background-color:#aef;`<br>
`    font-size: 123px;`<br>
`    background: -webkit-linear-gradient(top, #f9f9f9, #fff);`<br>
`}`<br>
`#Cab`<br>
`{`<br>
`    background-color: #ABC;`<br>
`    border: 2px dashed #fff;`<br>
`}  ` <br>

### Sample Output

> `#FfFdF8`<br>
`#aef`<br>
`#f9f9f9`<br>
`#fff`<br>
`#ABC`<br>
`#fff`<br>

### Explanation

#BED and #Cab satisfy the Hex Color Code criteria, but they are used as selectors and not as color codes in the given CSS.<br>

Hence, the valid color codes are:<br>

#FfFdF8<br>
#aef<br>
#f9f9f9<br>
#fff<br>
#ABC<br>
#fff<br>

**Note:** There are no comments ( // or /* */) in CSS Code.

```python
import re

N = int(input())

pattern = re.compile(r':?.(#[0-9a-fA-F]{6}|#[0-9a-fA-F]{3})')
for _ in range(N):
    matches = re.findall(pattern,input())
    if matches:
        print(*matches, sep="\n")
```
