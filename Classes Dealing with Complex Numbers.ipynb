# Classes: Dealing with Complex Numbers
For this challenge, you are given two complex numbers, and you have to print the result of their addition, subtraction, multiplication, division and modulus operations.<br>
The real and imaginary precision part should be correct up to two decimal places.

### Input Format
One line of input: The real and imaginary part of a number separated by a space.

### Output Format

For two complex numbers $C$ and $D$, the output should be in the following sequence on separate lines:<br>
* $C + D$
* $C - D$
* $C * D$
* $C / D$
* $mod(C)$
* $mod(D)$

For complex numbers with non-zero real $(A)$ and complex part $(B)$, the output should be in the following format:<br>
$A+Bi$<br>
Replace the plus symbol $(+)$ with a minus symbol $(-)$ when $B<0$.<br>
For complex numbers with a zero complex part i.e. real numbers, the output should be:<br>
$A+0.00i$<br>
For complex numbers where the real part is zero and the complex part $(B)$ is non-zero, the output should be:<br>
$0.00+Bi$

### Sample Input

> `2 1`<br>
`5 6`

### Sample Output

> `7.00+7.00i`<br>
`-3.00-5.00i`<br>
`4.00+17.00i`<br>
`0.26-0.11i`<br>
`2.24+0.00i`<br>
`7.81+0.00i`

### Concept

Python is a fully object-oriented language like C++, Java, etc. For reading about classes, refer here.<br>

Methods with a double underscore before and after their name are considered as built-in methods. They are used by interpreters and are generally used in the implementation of overloaded operators or other built-in functionality.<br>

>`__add__-> Can be overloaded for + operation`<br>

>`__sub__ -> Can be overloaded for - operation`<br>

>`__mul__ -> Can be overloaded for * operation`<br>

For more information on operator overloading in Python, refer [here](https://docs.python.org/3.2/reference/datamodel.html).

```python
import math

class Complex(object):
    def __init__(self, real, imaginary):
        self.real = real
        self.imaginary = imaginary
        
    def __add__(self, no):
        addition = complex(self.real, self.imaginary) + complex(no.real, no.imaginary)
        return Complex(addition.real, addition.imag).__str__()
        
    def __sub__(self, no):
        subtraction = complex(self.real, self.imaginary) - complex(no.real, no.imaginary)
        return Complex(subtraction.real, subtraction.imag).__str__()
        
    def __mul__(self, no):
        multiplication = complex(self.real, self.imaginary) * complex(no.real, no.imaginary)
        return Complex(multiplication.real, multiplication.imag).__str__()

    def __truediv__(self, no):
        division = complex(self.real, self.imaginary) / complex(no.real, no.imaginary)
        return Complex(division.real, division.imag).__str__()

    def mod(self):
        return Complex(abs(complex(self.real,self.imaginary)),0)

    def __str__(self):
        if self.imaginary == 0:
            result = "%.2f+0.00i" % (self.real)
        elif self.real == 0:
            if self.imaginary >= 0:
                result = "0.00+%.2fi" % (self.imaginary)
            else:
                result = "0.00-%.2fi" % (abs(self.imaginary))
        elif self.imaginary > 0:
            result = "%.2f+%.2fi" % (self.real, self.imaginary)
        else:
            result = "%.2f-%.2fi" % (self.real, abs(self.imaginary))
        return result

if __name__ == '__main__':
    c = map(float, input().split())
    d = map(float, input().split())
    x = Complex(*c)
    y = Complex(*d)
    print(*map(str, [x+y, x-y, x*y, x/y, x.mod(), y.mod()]), sep='\n')
```
