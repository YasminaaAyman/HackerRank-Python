# Zeros and Ones
### [zeros](https://numpy.org/doc/stable/reference/generated/numpy.zeros.html#numpy-zeros)

The zeros tool returns a new array with a given shape and type filled with $0$'s.<br>

> `import numpy`<br><br>
`print numpy.zeros((1,2))                    #Default type is float`<br>
`#Output : [[ 0.  0.]] `<br><br>
`print numpy.zeros((1,2), dtype = numpy.int) #Type changes to int`<br>
`#Output : [[0 0]]`<br>

### [ones](https://numpy.org/doc/stable/reference/generated/numpy.ones.html#numpy-ones)

The ones tool returns a new array with a given shape and type filled with $1$'s.<br>

> `import numpy`<br><br>
`print numpy.ones((1,2))                    #Default type is float`<br>
`#Output : [[ 1.  1.]]`<br><br> 
`print numpy.ones((1,2), dtype = numpy.int) #Type changes to int`<br>
`#Output : [[1 1]]`<br>
